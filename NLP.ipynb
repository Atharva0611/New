```python
import nltk

from nltk.stem import WordNetLemmatizer

# Download Dictionary

nltk.download('wordnet')

# Intailize the "Libriarian"

lemmatizer = WordNetLemmatizer()

words = ["rocks","corporate","better","running","was", "feet"]

print(f"{'Original':<12}|{'Lemmatized':<12}")

print("-"*30)

for word in words:

  # Default is usually noun

  lemma = lemmatizer.lemmatize(word, pos= "v") if word in ["was","running"] else lemmatizer.lemmatize(word)

  print(f"{word:<12}|{lemma:<12}")




```
